```python
#"The following code loads this library,
#defines two column vectors, ket0 and ket1,
#corresponding to the qubit state vectors
#∣0⟩ and ∣1⟩, and then prints their average."#

import numpy as np

ket0 = np.array([[1], [0]])
ket1 = np.array([[0], [1]])

print(ket0 / 2 + ket1 / 2)
```

```python
import numpy as np
ket0 = np.array([[1] ,[0]])
ket1 = np.array([[0] ,[1]])

print(ket0/2 + ket1/2)
```

```python
#"We can also use array to create matrices
#that can represent operations. "#

M1 = np.array([[1, 1], [0, 0]])
M2 = np.array([[1, 0], [0, 1]])
M = M1 / 2 + M2 / 2
print(M)
```

```python
M1 = np.array([[1, 1], [0, 0]])
M2 = np.array([[1, 0], [0, 1]])
M = M1/2 + M2/2
print(M)
```

```python
#'Matrix multiplication, including
#matrix-vector multiplication
#as a special case, can be
#performed using the @ operator.'#

ket0 = np.array([[1], [0]])
ket1 = np.array([[0], [1]])
M1 = np.array([[1, 1], [0, 0]])
M2 = np.array([[1, 0], [0, 1]])
M = M1/2 + M2/2

print(M1 @ ket1)
print(M1 @ M2)
print(M @ M)
```

```python
from qiskit.visualization import array_to_latex

display(array_to_latex(ket0))
print("\n")
display(array_to_latex(ket1))
print("\n")
display(array_to_latex(M1))
print("\n")
display(array_to_latex(M2))
print("\n")
display(array_to_latex(M))
print("\n")
display(array_to_latex(M1 @ ket1))
print("\n")
display(array_to_latex(M1 @ M2))
print("\n")
display(array_to_latex(M @ M))
```

```python
from qiskit.quantum_info import Statevector
from numpy import sqrt

u = Statevector([1/sqrt(2), 1/sqrt(2)])
v = Statevector([(1 + 2.0j)/3, - 2/3])
w = Statevector([1/3, 2/3])

#Statevector class include draw method
#for displaying Statevector in
#text, latex, and latex_source
#use print for latex_source
display(u.draw("text"))
display(u.draw("latex"))
print(u.draw("latex_source"))
```

```python
#Statevector also include is_valid method
#To check if a vector follows ||v||
display(u.is_valid())
display(v.is_valid())
display(w.is_valid())
```
