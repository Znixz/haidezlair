```python
from qiskit.quantum_info import Statevector, Operator
from qiskit import QuantumCircuit
```

```python

H = Operator.from_label("H")
S = Operator.from_label("S")
T = Operator.from_label("T")
HST_ = [H, S, T]
for operator in HST_:
    display(operator.draw("latex"))
```

```python
zero = Statevector.from_label("0")
circuit = QuantumCircuit(1)

circuit.h(0)
display(circuit.draw(output = "mpl"))
v = zero.evolve(circuit)
display(v.draw("latex"))

circuit.s(0)
display(circuit.draw(output = "mpl"))
v = zero.evolve(circuit)
display(v.draw("latex"))

circuit.h(0)
display(circuit.draw(output = "mpl"))
v = zero.evolve(circuit)
display(v.draw("latex"))

circuit.t(0)

display(circuit.draw(output = "mpl"))
v = zero.evolve(circuit)
display(v.draw("latex"))
```

```python
H = Operator.from_label("H")
display(H.draw("latex"))
```

```python
outcome, state = v.measure()
print(f"Measured: {outcome}\n State after measurement:")
display(state.draw("latex"))
```

```python
one = Statevector.from_label("1")
ket1 = one.evolve(circuit)

display(circuit.draw(output = "mpl"))
display(ket1.draw("latex"))
```

```python
outcome, state = ket1.measure()
print(f"Measured: {outcome}\nState after measurement:")
display(state.draw("latex"))
```

```python
from qiskit.visualization import plot_histogram

v_sample = v.sample_counts(1000)
plot_histogram(v_sample)
```

```python
ket1_sample = ket1.sample_counts(1000)
plot_histogram(ket1_sample)
```

```python
circuit = QuantumCircuit(2)
circuit.h(0)
circuit.cx(0, 1)
display(circuit.draw(output = "mpl"))
```

```python
display(Operator(circuit).draw("latex"))
```

```python
q = Statevector.from_label("00")
r= q.evolve(circuit)
display(r.draw("latex"))
```

```python
q_sample = r.sample_counts(100)
plot_histogram(q_sample)
```
