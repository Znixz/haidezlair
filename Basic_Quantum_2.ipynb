```python
#The Statevector class has a tensor() method,
#which returns the tensor product of a Statevector
#with another Statevector given as an argument.

#Example:
#Create two Statevector ketzero and ketone.

zero = Statevector.from_label("0")
one = Statevector.from_label("1")

#use the tensor() method to create another
#new vector name psi,
#psi is the tensor product of zero
#and one given as argument

psi = zero.tensor(one)

#display the result in latex

display(psi.draw("latex"))
```

```python
plus = Statevector.from_label("+")
minus = Statevector.from_label("-")
plus_i = Statevector.from_label("r")
minus_i = Statevector.from_label("l") #small letter of L

group_id = {"plus(+)": plus, "minus(-)": minus, "plus_i(+i)(r)": plus_i, "minus_i(-i)(l)": minus_i}

for i in group_id:
    print(f"\n{i}:")
    display(group_id[i].draw("latex"))
```

```python
phi = plus.tensor(minus_i)

display(phi.draw("latex"))

#Note that ^ can be use to represent
#.tensor() method which output same thing

display((plus ^ minus_i).draw("latex"))
```

```python
#The Operator class also has a tensor method
#as well as a from_label method.

#Example:

H = Operator.from_label("H")
Id = Operator.from_label("I")
X = Operator.from_label("X")

group_id_1 = {"H": H, "Id": Id, "X": X}

for i in group_id_1:
    print(f"\n{i}:")
    display(group_id_1[i].draw("latex"))
```

```python
#Tensor Operations with Operators

display(H.tensor(Id).draw("latex"))
display((H ^ Id).draw("latex"))
print("\n")
display(H.tensor(Id).tensor(X).draw("latex"))
display((H ^ Id ^ X).draw("latex"))
```
