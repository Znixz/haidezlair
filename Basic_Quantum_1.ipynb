display(T.draw("latex"))

```python
#Working with Operator and Statevector

Y = Operator([[0, -1.0j], [1.0j, 0]])
H = Operator([[1/sqrt(2), 1/sqrt(2)], [1/sqrt(2), -1/sqrt(2)]])
S = Operator([[1, 0], [0, 1.0j]])
T = Operator([[1, 0], [0, (1 + 1.0j)/sqrt(2)]])
X = Operator([[0, 1], [1, 0]])
Operators = [Y, H, S, T, X]
display(Operators[0].draw("latex"))
for operator in Operators:
    display(operator.draw("latex"))
```

```python
#Working with Operator and single Basis State 0
#Using evolve method

v = Statevector([1, 0])
statistics_1 = v.sample_counts(1000)
print("Stats before v evolve:")
display(v.draw("latex"))
plot_histogram(statistics_1)

v = v.evolve(H)
v = v.evolve(T)
v = v.evolve(H)
v = v.evolve(S)
v = v.evolve(Y)

print("Stats after gate effect:")
display(v.draw("latex"))
statistics_2 = v.sample_counts(1000)
plot_histogram(statistics_2)
```
